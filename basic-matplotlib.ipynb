#Learning Matplotlib

```python
# Matplotlib is used for data visualization
# In this case the data is created using numpy
import numpy as np
import matplotlib.pyplot as plt
```

##1.Preliminary

```python
x = np.array([1, 2, 3, 4 , 5])
y = np.array([6, 7, 8, 9 , 10])
z = np.array([2, 4, 6, 8 , 10])
```

```python
# To simply plot the data, use plot()
# Multiple use of plot() can cause many ploting on graph
plt.plot(x, y)
plt.plot(x, z)
plt.show()
```

##2. Color and Marker

```python
# Create sine wave generator

def sine_generator(amplitude, frequency, finish_time, theta):
  time =  np.arange(0, finish_time, 0.1)
  y = amplitude * np.sin(2*frequency*time + np.deg2rad(theta))
  return time, y
```

```python
# To plot a graph with specific color and style
# I can assign intial name of color with the symbol 
# of line style as 3rd arguments to plot()

x1, y1 = sine_generator(1, 1, 4, 0)
plt.plot(x1, y1)

x2, y2 = sine_generator(1, 1, 4, 30)
plt.plot(x2, y2, 'r')

x3, y3 = sine_generator(1, 1, 4, 45)
plt.plot(x3, y3, 'b--')

x4, y4 = sine_generator(1, 1, 4, 60)
plt.plot(x4, y4, 'g.')

x5, y5 = sine_generator(1, 1, 4, 90)
plt.plot(x5, y5, 'c-o')

plt.show()
```

```python
# Also to set color and marker of graph i can use setp

plot1 = plt.plot(x1, y1)
plt.setp(plot1, color='y')

plot2 = plt.plot(x2, y2)
plt.setp(plot2, color='r')

plot3 = plt.plot(x3, y3)
plt.setp(plot3, color='b', linestyle='--')

plot4 = plt.plot(x4, y4)
plt.setp(plot4, color='g', linestyle='dotted')

plot5 = plt.plot(x5, y5)
plt.setp(plot5, color='c', linestyle='-.', linewidth=4)

plt.show()
```

##3. Set Axis

```python
# To set the axis of graph, I can use axis()
# with list of specified size of axis
# the list is [xmin, xmax, ymin, ymax]

x, y = sine_generator(1, 1, 8, 0)

plt.plot(x, y)

plt.axis([0, 8, -4, 4])

plt.show()
```

##4. Label and Title

```python
# To clarify what graph and what data in x and y axis
# I can use title() and x/ylabel() to give it name
plt.plot(x, y)

plt.title("Sine Graph")

plt.xlabel("Time (second)")
plt.ylabel("Amplitude (cm)")

plt.show()
```

##5. Legend

```python
x1, y1 =  sine_generator(1, 1, 8, 0)

x2, y2, = sine_generator(1, 1, 8, 90)

plt.plot(x1, y1, label="sin(0)")
plt.plot(x2, y2, label="sin(90)")
plt.legend()

plt.show()
```

```python
plt.figure(1)
ax = plt.subplot(333)
plt.plot(x1, y1, label="sin(0)")
plt.plot(x2, y2, label="sin(90)")
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width-0.7, box.height])
plt.legend(loc="upper center", bbox_to_anchor=(1.2,1))

plt.show()
```
